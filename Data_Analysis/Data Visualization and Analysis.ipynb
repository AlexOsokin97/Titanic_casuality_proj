```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_first_class = pd.read_csv('pclass1_data_.csv')
df_second_class = pd.read_csv('pclass2_data_.csv')
df_third_class = pd.read_csv('pclass3_data_.csv')
```

```python
df_first_class.describe()
```

```python
df_second_class.describe()
```

```python
df_third_class.describe()
```

```python
CV1 = df_first_class['Age'].std()/df_first_class['Age'].median()
CV2 = df_second_class['Age'].std()/df_second_class['Age'].median()
CV3 = df_third_class['Age'].std()/df_third_class['Age'].median()
```

```python
print('coefficient of variation of ages in the first class: ', CV1)
print('coefficient of variation of ages in the second class: ', CV2)
print('coefficient of variation of ages in the third class: ', CV3)
```

## By calculting the coefficient of variation of the ages of all the three classes we can see that the ages don't have high varation which helps us to assume that the data is well distributed and there are not many outliers that can affect our evaluation

```python
f, axes = plt.subplots(1, 3)
f.set_size_inches(18,5)

sns.distplot(df_first_class['Age'], ax=axes[0])
axes[0].set_title('Age % of first class')
axes[0].set_ylabel('Precent')

sns.distplot(df_second_class['Age'], ax=axes[1])
axes[1].set_title('Age % of second class')
axes[1].set_ylabel('Precent')

sns.distplot(df_third_class['Age'], ax=axes[2])
axes[2].set_title('Age % of third class')
axes[2].set_ylabel('Precent')
```

## As we can see according to the distribution plots our first class age precentage is well distributed while second and third class age precentage is very close to being well distributed. As a result I may apply regularization.

```python
f1, axes1 = plt.subplots(1, 3)
f1.set_size_inches(20,10)

palette1 = {'male':'#e74c3c', 'female':'#2ecc71'}

sns.stripplot(x="Survived", y="Age",hue="Sex",palette=palette1, hue_order=['male','female'],
            data=df_first_class, ax=axes1[0])
axes1[0].set_title("First Class Died(0)/Survived(1)")
axes1[0].legend(ncol=2, loc="upper right", frameon=True)

sns.stripplot(x="Survived", y="Age",hue="Sex",palette=palette1, hue_order=['male','female'],
            data=df_second_class, ax=axes1[1])
axes1[1].set_title("Second Class Died(0)/Survived(1)")
axes1[1].legend(ncol=2, loc="upper right", frameon=True)

sns.stripplot(x="Survived", y="Age",hue="Sex",palette=palette1, hue_order=['male','female'],
            data=df_third_class, ax=axes1[2])
axes1[2].set_title("Third Class Died(0)/Survived(1)")
axes1[2].legend(ncol=2, loc="upper right", frameon=True)
```

## Woah! By plotting Age and Survived in a catplot and using different colors for Males and Females passangers we can clearly see that not only the mortality rate rises as the passanger class gets lower, but also most of the casualties are males between their 20s - 40s in each of the passanger class

#### (1 - survived, 0 - died)

```python
f2, axes2 = plt.subplots(3,2)
f2.set_size_inches(13,20)
sns.swarmplot(x="SibSp", y="Age",
             data=df_first_class,hue='Survived', ax=axes2[0][0])
axes2[0][0].set_xlabel("Siblings/Spouces")
axes2[0][0].set_title("First Class Age to number of Siblings/Spouces relation")

sns.swarmplot(x="Parch", y="Age",
              data=df_first_class,hue='Survived', ax=axes2[0][1])
axes2[0][1].set_xlabel("Parents/Children")
axes2[0][1].set_title("First Class Age to number of Parents/Children relation")

##############################################################################################################################

sns.swarmplot(x="SibSp", y="Age",
             data=df_second_class,hue='Survived', ax=axes2[1][0])
axes2[1][0].set_xlabel("Siblings/Spouces")
axes2[1][0].set_title("Second Class Age to number of Siblings/Spouces relation")

sns.swarmplot(x="Parch", y="Age",
              data=df_second_class,hue='Survived', ax=axes2[1][1])
axes2[1][1].set_xlabel("Parents/Children")
axes2[1][1].set_title("Second Class Age to number of Parents/Children relation")

##############################################################################################################################

sns.swarmplot(x="SibSp", y="Age",
             data=df_third_class,hue='Survived', ax=axes2[2][0])
axes2[2][0].set_xlabel("Siblings/Spouces")
axes2[2][0].set_title("Third Class Age to number of Siblings/Spouces relation")

sns.swarmplot(x="Parch", y="Age",
              data=df_third_class,hue='Survived', ax=axes2[2][1])
axes2[2][1].set_xlabel("Parents/Children")
axes2[2][1].set_title("Third Class Age to number of Parents/Children relation")
```

### As we can see from the these plots when it comes to 'Age to siblings/spouces relation' we can clearly see that in the first and second class most of it's passangers were lone travelers and had at most 1 spouce (more likely passangers in their 20s - 60s) or 1 sibling (most likely passangers that were infants or kids under 20). The same we can see in the 'Age to parents/children relation' plot where most children under 20 were both with their parents and those over 20 had 1 - 2 children. 
### As we can see in the third class plots many passangers were also lone travelers But, alot of them had much more siblings/spouces(up to 8 siblings/spouces) or parents/children(up to 6 siblings/spouces).

### Surprisingly enough, most of the first and second class passangers who had siblings/spouces and parents/children survived, while most of the passangers who had siblings/spouces and parents/children did not survive.

### We can see from the plot that most females under the age of 20 are from Cherbourg and traveled in the second and third. Males from Queenstown were the oldest in every passanger class. Unsurprisingly, the first class papulation both females and males are on average all above 30 for females while males were mostly above 40. In the third class though, we can see that most of it's passangers do not even reach their 25s-30s.

```python
cp = sns.countplot(x='Survived', hue='Sex', data=titanic_df, palette='Set1')
```

```python
titanic_df['Sex'] = titanic_df['Sex'].apply(lambda x: 1 if x == 'female' else 0)
```

```python
corr = titanic_df.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
h = sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
h.set_title('Correlation Heatmap')
fig1 = h.get_figure()
fig1.savefig('corrHeatmap.png')
```

### As we can see according to the heatmap there's a strong correlation between the passengers sex and his/hers survival. In additon, the passenger's fare also has pretty strong fare on his survival.

```python
titanic_df['Sex'] = titanic_df['Sex'].apply(lambda x: 'female' if x == 1 else 'male')
```

```python
f5, axes5 = plt.subplots(1, 1)
f5.set_size_inches(12,7)

c = sns.stripplot(x="Survived", y="Age",hue="Sex",palette=palette1, hue_order=['male','female'],
            data=titanic_df, ax=axes5)
c.set_title('Male / Female casualties and survivers')
fig2 = c.get_figure()
fig2.savefig('MaleFemaleSurvived.png')
```

### As we can see from the whole dataset most of titanics's casualties are men between the age of 20 and 40 while most of tittanic's survivors are females between 20 and 30. But, lets not forget the ratio between males and females who were onboard  2:1 (for every 2 males there was 1 female) which may be one of the causes that there were more male casualties than female casualties.

```python
grid = sns.FacetGrid(titanic_df, col='Survived', row='Pclass', height=3.2, aspect=2.2)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend();
grid.savefig('grid.png')
```

### As we can see from this grid most of the casualties were from the third class while in the other two classes the amount of casualties was pretty low. The reason for this might be the room locations of the travel classes. The third class rooms were mostly located at the lower part of the titanic while 2nd and 1st class rooms were located at the upper level of the titanic. As a result, 1st and 2nd class passengers had more time to get to the life boats which were located at the deck of the ship and hence survive.

```python
survived = 'survived'
not_survived = 'not survived'
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(10, 4))
women = titanic_df[titanic_df['Sex']=='female']
men = titanic_df[titanic_df['Sex']=='male']
ax = sns.distplot(women[women['Survived']==1].Age.dropna(), bins=18, label = survived, ax = axes[0], kde =False)
ax = sns.distplot(women[women['Survived']==0].Age.dropna(), bins=40, label = not_survived, ax = axes[0], kde =False)
ax.legend()
ax.set_title('Female')
ax = sns.distplot(men[men['Survived']==1].Age.dropna(), bins=18, label = survived, ax = axes[1], kde = False)
ax = sns.distplot(men[men['Survived']==0].Age.dropna(), bins=40, label = not_survived, ax = axes[1], kde = False)
ax.legend()
_ = ax.set_title('Male')
fig.savefig('fig.png')
```
